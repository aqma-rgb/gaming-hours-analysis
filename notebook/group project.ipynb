```python
plt.figure(figsize=(8,5))
sns.histplot(df['daily_gaming_hours'], bins=20, kde=True)
plt.title("Distribution of Daily Gaming Hours")
plt.xlabel("Daily Gaming Hours")
plt.ylabel("Count")
plt.show()

```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['academic_or_work_score'], bins=20, kde=True)
plt.title("Distribution of Academic / Work Score")
plt.xlabel("Score")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='daily_gaming_hours',
    y='academic_or_work_score',
    data=df
)
plt.title("Daily Gaming Hours vs Academic / Work Score")
plt.xlabel("Daily Gaming Hours")
plt.ylabel("Academic / Work Score")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(
    df[['daily_gaming_hours',
        'weekly_gaming_hours',
        'sleep_hours',
        'stress_level',
        'focus_level',
        'academic_or_work_score',
        'productivity_level']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Heatmap of Gaming & Performance Factors")
plt.show()
```

```python
bins = [0, 2, 4, 6, 10]
labels = ['Low', 'Moderate', 'High', 'Very High']

df['gaming_level'] = pd.cut(
    df['daily_gaming_hours'],
    bins=bins,
    labels=labels
)

df.groupby('gaming_level')['academic_or_work_score'].mean()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x='gaming_level',
    y='academic_or_work_score',
    data=df
)
plt.title("Average Academic / Work Score by Gaming Level")
plt.xlabel("Gaming Level")
plt.ylabel("Average Score")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(10,6))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()
```

```python
selected_cols = [
    'daily_gaming_hours',
    'weekly_gaming_hours',
    'sleep_hours',
    'stress_level',
    'focus_level',
    'academic_or_work_score',
    'productivity_level'
]

plt.figure(figsize=(10,6))
sns.heatmap(
    df[selected_cols].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Between Gaming, Health, and Performance")
plt.show()
```
